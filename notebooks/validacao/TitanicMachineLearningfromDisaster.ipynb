```python
# Conta quantos passageiros são homens e quantos são mulheres, permitindo uma primeira análise exploratória da variável Sex.
train['Sex'].value_counts() # 577 homens e 314 mulheres
```

```python
'''
 Cria uma função para transformar a variável categórica Sex em numérica (binária):

 female → 1

 male → 0
'''

def transformar_sexo(valor):
    if valor == 'female':
        return 1
    else:
        return 0

train['Sex_binario'] = train['Sex'].map(transformar_sexo)
```

```python
train.head()
```

```python
variaveis = ['Sex_binario', 'Age'] # Famale = 1 - Male = 0
```

```python
'''
 Separa os dados em:

 X: variáveis independentes (features usadas para prever).

 y: variável alvo (Survived), que indica se o passageiro sobreviveu ou não.
'''

X = train[variaveis]
y = train['Survived']
```

```python
X.head()
```

```python
y.head()
```

```python
# Comentário: Substitui valores ausentes (NaN) nas variáveis preditoras por -1.

X = X.fillna(-1)
```

```python
# Treina o modelo Random Forest com os dados de treino (X, y).

modelo.fit(X, y)
```

```python
# Seleciona, no dataset de teste, as mesmas variáveis utilizadas no treinamento para fazer previsões.

X_prev = test[variaveis]
```

```python
'''
Transforma a variável Sex em binária também no conjunto de teste, 
preenche valores ausentes com -1 e visualiza as primeiras linhas preparadas para predição.
'''

test['Sex_binario'] = test['Sex'].map(transformar_sexo)
X_prev = X_prev.fillna(-1)
X_prev.head()
```
